```python
#Este codigo muestra todas las boxes dentro de una imagen. Luego será usado nuevamente
#test_img_id=['3c095a6526719d58']       # este es para probar con la araña
img_url =img_box_pd.loc[img_box_pd["ImageID"]==test_img_id[0]]['OriginalURL'].values[0]
img = io.imread(img_url)
h,w,ch =img.shape
bbox = anota_box_pd[anota_box_pd['ImageID']==test_img_id[0]]
for i,f in bbox.iterrows():
  print('indice:{}'.format(i))
  print('fes:{}'.format(f))
  xmin =int(f['XMin']*w)    # des normalizar los datos X y Y
  xmax = int(f['XMax']*w)
  ymin = int(f['YMin']*h)
  ymax = int(f['YMax']*h)
  label = f['LabelName']
  clase = clas_desc_pd[clas_desc_pd[0]==label]
  cv2.rectangle(img, (xmin,ymin), (xmax,ymax), (255,0,0), 5)
  font = cv2.FONT_HERSHEY_SIMPLEX
  cv2.putText(img, clase[1].values[0], (xmin,ymin-10), font, 3, (0,255,0), 5)
plt.figure(figsize=(15,10))
plt.title('Image with Bounding Box')
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
#Buscar en la tabla de descripciones de clases los ID de Personas y Carros
clas_desc_pd.loc[clas_desc_pd[1].isin(['Elephant','Owl'])]
```

```python
#Tomemos aleatoriamente 100 imagenes de personas y 100 de carros
#Lineas de clases
weapon_pd =clas_desc_pd[clas_desc_pd[1]=='Elephant']
toy_pd = clas_desc_pd[clas_desc_pd[1]=='Owl']
#Etiquetas de clases (Estas etiquetas serán primordiales cuando se vaya a copiar a disco local)
weapon_label_pd = weapon_pd[0].values[0]
toy_label_pd = toy_pd[0].values[0]
#Usemos las etiquetas para bajar las box de anotaciones
weapon_box_an= anota_box_pd[anota_box_pd['LabelName']==weapon_label_pd]
toy_box_an = anota_box_pd[anota_box_pd['LabelName']==toy_label_pd]
#Bajemos los ID de las clases
weapon_id=weapon_box_an['ImageID']
toy_id =toy_box_an['ImageID']
#Dado que pueden haber varias personas en una imagen, podrían aparecer ID's repetidos. Vamos a usar una sola
weapon_id = np.unique(weapon_id)
toy_id= np.unique(toy_id)
#******Escoger aleatoriamente 1000 imagenes de cada clase
n=1000
sub_weapon_id =random.sample(list(weapon_id),n)
sub_toy_id = random.sample(list(toy_id),n)
```

```python
#Usemos pandas para crear los dos conjuntos de datos a partir de los Sub ID's
sub_weapon_pd = img_box_pd.loc[img_box_pd['ImageID'].isin(sub_weapon_id)]
sub_toy_pd = img_box_pd.loc[img_box_pd['ImageID'].isin(sub_toy_id)]
print(sub_weapon_pd.shape)
print(sub_toy_pd.shape)
```

```python
"""Este codigo nos permite tomar el 80% de las 200 imagenes (carros y personas) y
copiarlas a la carpeta train. El 20% restante copiadas a la carpeta test."""
clas=['Elephant', 'Owl']
for i in range(len(clas)):
  imgs = os.listdir(clas[i])
  random.shuffle(imgs)
  percent = int(n*0.8)
  train_imgs=imgs[:percent]     # de 0% a 80%
  test_imgs = imgs[percent:]    # de 80% a 100%

  for f in range(len(train_imgs)):
    src = os.path.join(clas[i],train_imgs[f])
    dst = os.path.join('/content/drive/MyDrive/Colab Notebooks/VISION COMPUTACIONAL/LABORATORIO_2/dataset/train2022/'+train_imgs[f])
    copyfile(src,dst)
  for f in range(len(test_imgs)):
    src = os.path.join(clas[i],test_imgs[f])
    dst = os.path.join('/content/drive/MyDrive/Colab Notebooks/VISION COMPUTACIONAL/LABORATORIO_2/dataset/test2022/'+test_imgs[f])
    copyfile(src,dst)
#print(imgs)
```

```python
import sys
```

```python
"""Aqui creamos un dataframe de pandas con la información
que es necesaria para nuestro entrenamiento: nombre del archivo imagen,
x1,x2,y1,y2 (datos de la box) y la etiqueta de clase (Persona o Carro).""" 
etiquetas =[weapon_label_pd, toy_label_pd]
train_df = pd.DataFrame(columns=['Nombre','Xmin','Xmax','Ymin', 'Ymax','Label'])

train_imgs = os.listdir('/content/drive/MyDrive/Colab Notebooks/VISION COMPUTACIONAL/LABORATORIO_2/dataset/train2022/')
for i in range(len(train_imgs)):
  sys.stdout.flush()
  img_nombre = train_imgs[i]
  img_id=img_nombre[0:16]       # aqui se coge el nombre del archivo, cantidad de caracteres del nombre de la imagen
  tmp_df = anota_box_pd[anota_box_pd['ImageID']==img_id]
  for index, row in tmp_df.iterrows():
    labelName = row['LabelName']
    for i in range(len(etiquetas)):
      if labelName == etiquetas[i]:
        train_df = train_df.append({'Nombre':img_nombre,
                                    'Xmin':row['XMin'],
                                    'Xmax':row['XMax'],
                                    'Ymin':row['YMin'],
                                    'Ymax':row['YMax'],
                                    'Label':clas[i]},
                                   ignore_index=True)

#SE DEMORA ALREDEDOR DE 22 MINUTOS
```

```python
"""Este codigo nos sirve para crear un archivo texto donde estara la informacion
de cada imagen de entrenamiento con sus datos de bbox y etiqueta.
Observe como cada linea es un bbox diferente. El archivo de imagen se puede repetir
tantas veces como bbox tenga la imagen."""
train_df = pd.read_csv('train.csv')
with open("/content/drive/MyDrive/Colab Notebooks/VISION COMPUTACIONAL/LABORATORIO_2/entrenamiento2022.txt", 'w+') as f:
  for idx, row in train_df.iterrows():
    img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/VISION COMPUTACIONAL/LABORATORIO_2/dataset/train2022/'+ row['Nombre'])
    h,w = img.shape[:2]
    x1 = int(row['Xmin']*w)
    x2 = int(row['Xmax']*w)
    y1 = int(row['Ymin']*h)
    y2 = int(row['Ymax']*h)
    archivo = os.path.join('/content/drive/MyDrive/Colab Notebooks/VISION COMPUTACIONAL/LABORATORIO_2/dataset/train2022',row['Nombre'])
    label = row['Label']
    f.write(archivo + ',' + str(x1) + ',' + str(y1) + ',' + str(x2) + ',' + str(y2) + ',' + label + '\n')

# 8 MINUTOS
```
